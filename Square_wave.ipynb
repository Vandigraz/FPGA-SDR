```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f = 1  # 1 Hz
t = np.linspace(0, 1, 1000)
A=1
```

```python
harmonic_1 = A * np.sin(2 * np.pi * f * t)
harmonic_3 = (A / 3) * np.sin(2 * np.pi * 3 * f * t)
harmonic_5 = (A / 5) * np.sin(2 * np.pi * 5 * f * t)
harmonic_7 = (A / 7) * np.sin(2 * np.pi * 7 * f * t)
harmonic_9 = (A / 9) * np.sin(2 * np.pi * 9 * f * t)
harmonic_11 = (A / 11) * np.sin(2 * np.pi * 11 * f * t)
harmonic_13 = (A / 13) * np.sin(2 * np.pi * 13 * f * t)
harmonic_15 = (A / 15) * np.sin(2 * np.pi * 15 * f * t)
```

```python
square_wave_approx = harmonic_1 + harmonic_3 + harmonic_5 + harmonic_7 + harmonic_9 + harmonic_11 + harmonic_13 + harmonic_15
```

```python
sns.kdeplot(data= square_wave_approx)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(t, square_wave_approx, label="Approximated Square Wave")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Square Wave Approximation using Harmonics")
plt.legend()
plt.grid(True)
plt.show()
```

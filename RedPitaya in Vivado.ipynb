## REDPITAYA AND VIVADO

This is a documentation on how you can accurately programme the SDR 122-16 using VIVADO.

#### Things to note that are mandatory

1. The board only uses vivado 2020.1 version. I shall explain why in the later cells
2. I intend to illustrate while using Windows 11.
3. You need to have a clock running the board at precisely 122.88MHz
4. You need a MicroUSB cable so as to connect it with your Computer.
5. As we are using windows 11, you need to have a Linux environment such as WSL which is downloaded from Microsoft website directly.
6. The above are just but a summary of what is in the guidelines from REDPITAYA site

#### Reasons to use vivado 2020.1

1. According to RedPitaya, Red Pitaya software was written for Vivado version 2020.1.
2. We shall have to use Vivado HLS 2020.1 which probably the newer versions might not have to launch the various attributed files for RedPitaya FPGA ecosystem.
   

### The use of the various terminal.


#### 1 Vivado HLS 2020.1

.This is probably the right and effective way to launch Vivado environment with the RedPitaya Ecosystem.


#### 2. Command prompt (CMD)

The command prompt, not necessarily the powershell, is used for two tasks.
1. Navigate to where the bit stream, generated by vivado is stored.
2. Sending the bit file to the board via ssh.
   ##### At this terminal, it was favourable to use the IP address of the board rather that its MAC address

3. It is efficient to know the IP address of your board before hand. The Mac address is clearly visible
   on the Ethernet port.

### 3. WSL 

The windows subsystem for linux does a tremendoes job in implementation/actualizing the file bit that was previously transferred via the CMD.

At this point I used the SSH with the MAD address. I guess you can also use the IP address.
#### ssh root@rp-f0*****.local to enable connectivity

use the ls command.
